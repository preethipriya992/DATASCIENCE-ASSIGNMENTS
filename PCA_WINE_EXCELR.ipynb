## PCA_Wine

 - Perform Principal component analysis and perform clustering using first 
3 principal component scores (both heirarchial and k mean clustering(scree plot or elbow curve) and obtain 
optimum number of clusters and check whether we have obtained same number of clusters with the original data 
(class column we have ignored at the begining who shows it has 3 clusters)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
data=pd.read_csv("wine.csv")
data
```

```python
data.info()
```

```python
data['Type'].value_counts()
```

```python
data2=data.iloc[:,1:]
data2
```

```python
#Data to numpy array and normalizing
data_array=data2.values
data_array
data_norm=scale(data_array)
data_norm
```

```python
# PCA
pca=PCA(n_components=13)           #13 numerical dimension
pca

data_pca=pca.fit_transform(data_norm)            # PC scores
data_pca
```

```python
pca.components_                # Covariance Matrix
```

```python
var=pca.explained_variance_ratio_             #Amount of variance/information each PCA contains
var
```

```python
var1=np.cumsum(np.round(var,4)*100)                #Cummulative sum of variance
var1
```

```python
plt.plot(var1)        #Variance Plot fro PC obtained
```

```python
f_df=pd.concat([data['Type'],pd.DataFrame(data_pca[:,:3],columns=['pc1','pc2','pc3'])],axis=1)
f_df
```

```python
df2=f_df.iloc[:,1:]
fig=plt.figure(figsize=(10,7))
sns.scatterplot(data=df2)
```

```python
fig=plt.figure(figsize=(10,7))
sns.scatterplot(data=f_df,x='pc1',y='pc2',hue='Type')
```

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering as agc
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(data_norm,'complete'))
```

```python
# Create 3 cluster
h_cluster=agc(n_clusters=3,affinity='euclidean',linkage='complete').fit(data_norm)
h_cluster
```

```python
h_cid=pd.Series(h_cluster.labels_)          
h_cid

# or y=pd.DataFrame(h_cluster.fit_predict(data_norm),columns=['clustersid'])   instead of above both
```

```python
data['clust_id']=h_cid
data
```

```python
data['clust_id'].value_counts()
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
'''We have normalised data. 
Elbow curve to find optimum K value
KMean cluster algorithm chooses centroids that minimizes the inertia/ within cluster sum of squares'''
```

```python
#WCSS
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data_norm)
    wcss.append(kmeans.inertia_)
    
```

```python
# Plot elbow curve
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# K=3
k_cluster=KMeans(3).fit(data_norm)
k_cluster.labels_
```

```python
data['k_id']=k_cluster.labels_
data
```
