## Association Rules_Book

Prepare rules for the all the data sets 
 - 1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values
 - 2) Change the minimum length in apriori algorithm
 - 3) Visulize the obtained rules using different plots 


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
data=pd.read_csv("book.csv")
data
```

```python
# Since the data is already in transaction format we need not preprocess

# APRIORI Algorithm, Association Rules
# 10% support and 70% confidence

#10% support
frequent_itemsets=apriori(data,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
# 70% confidence and sort lift ration in descending order
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=0.7)
rules
rules=rules.sort_values("lift", ascending=False)
rules
pd.set_option('display.max_rows',None)
rules
```

```python
# Here we can select association rules: 28, 77, 84, 95
# Leverage value of 0 indicates independence
# Hingh conviction value indicates consequent is highly depending on antecedent
```

```python
plt.scatter(rules["support"],rules["confidence"])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
