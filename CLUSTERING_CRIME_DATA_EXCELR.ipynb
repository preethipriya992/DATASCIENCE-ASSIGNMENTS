## Clustering_crime_data

Perform Clustering(Hierarchical, Kmeans & DBSCAN) for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
 - Murder -- Muder rates in different places of United States
 - Assualt- Assualt rate in different places of United States
 - UrbanPop - urban population in different places of United States
 - Rape - Rape rate in different places of United States

## Hierarchical Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering as agc
import scipy.cluster.hierarchy as sch
from sklearn.cluster import DBSCAN
```

```python
data=pd.read_csv("crime_data.csv")
data
```

```python
data.info()
```

```python
#Normalize data
def norm_fn(i):
    x=(i-i.min())/(i.max()-i.min())
    return x
```

```python
df_norm=norm_fn(data.iloc[:,1:])
df_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
#here we decide the clusters are 3
hc=agc(n_clusters=3,affinity='euclidean',linkage='complete')
hc
```

```python
#Predict cluster for observations
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
data['h_cid']=hc.labels_
data
```

```python
data.groupby('h_cid').agg(['mean']).reset_index()
```

## KMeans Clustering

```python
from sklearn.cluster import KMeans
```

```python
data=pd.read_csv("crime_data.csv")

def norm_fn(i):
    x=(i-i.min())/(i.max()-i.min())
    return x

df_norm=norm_fn(data.iloc[:,1:])
```

```python
# Elbow curve to find K value
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
```

```python
# from above scree plot we can select k=4
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
data['k_cid']=x
data
```

```python
data.iloc[:,1:5].groupby(data.k_cid).mean()

# same as    data.groupby('k_cid').agg(['mean']).reset_index()
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("crime_data.csv")
data.info()

df=data.iloc[:,1:5]
df
```

```python
df.values
```

```python
StdScaler=StandardScaler().fit(df.values)
StdScaler
```

```python
x=StdScaler.transform(df.values)
x
```

```python
from sklearn.neighbors import NearestNeighbors
neigh=NearestNeighbors(n_neighbors=8)         #2*4feature
nbrs=neigh.fit(x)
neigh
nbrs
distances, indices=nbrs.kneighbors(x)
distances
```

```python
distances=np.sort(distances,axis=0)
distances=distances[:,1]
plt.plot(distances)
```

```python
dbscan=DBSCAN(eps=1.25,min_samples=5)            #1+4features
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
dc=pd.DataFrame(dbscan.labels_,columns=['d_cid'])
dc
```

```python
pd.concat([data,dc],axis=1)
```
